```python
from PIL import Image
wine_mask = np.array(Image.open("head.png"))

```

```python
# https://stackoverflow.com/questions/43043263/word-cloud-in-python-with-customised-colour
    
def random_color_func(word=None, font_size=None, position=None,  orientation=None, font_path=None, random_state=None):
    h = int(360.0 * 21.0 / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(60, 120)) / 255.0)
    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
wc = WordCloud(width=1000, height=1000, margin=0, background_color="white", color_func=random_color_func, max_font_size = 200, max_words=200, mask=wine_mask, contour_width=1, contour_color='steelblue', collocations=False)

# generate word cloud
wc.generate(df2list)

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()

plt.show()
```

```python
# https://stackoverflow.com/questions/52910187/how-to-make-a-polygon-radar-spider-chart-in-python

assetdf = dataset.loc[dataset.features == 'asset classes']
labels=list(assetdf['xaxis'])
markers = [0, 1, 2, 3, 4, 5]
str_markers = ["0", "1", "2", "3", "4", "5"]
plt.rcParams['font.size'] = 16
def make_radar_chart(name, stats, attribute_labels = labels, plot_markers = markers, plot_str_markers = str_markers):

    labels = np.array(attribute_labels)

    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
    stats = np.concatenate((stats,[stats[0]]))
    angles = np.concatenate((angles,[angles[0]]))

    fig= plt.figure()
    ax = fig.add_subplot(111, polar=True)
    ax.plot(angles, stats, 'o-', linewidth=2)
    ax.fill(angles, stats, alpha=0.25)
    ax.set_thetagrids(angles * 180/np.pi, labels)
    plt.yticks(markers)
    #ax.set_title(name)
    ax.grid(True)

    fig.savefig("%s.png" % name)

    return plt.show()

make_radar_chart("asset classes", list(assetdf['yrs'])) # example
```
